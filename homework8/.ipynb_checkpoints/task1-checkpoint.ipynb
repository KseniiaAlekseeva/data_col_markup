```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from scipy import stats
from datetime import datetime
```

```python
df=pd.read_csv('train.csv')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
# обработка отсутств значений
numeric_cols = df.select_dtypes(include=[np.number]) # выбор числовых столбцов
df[numeric_cols.columns] = numeric_cols.fillna(numeric_cols.mean())
categorical_cols = df.select_dtypes(include=['object']) # выбор категориальных колонок
df[categorical_cols.columns] = categorical_cols.fillna(categorical_cols.mode().iloc[0])
df.drop_duplicates(inplace=True)
df.info()
```

```python
# анализ нечисловых столбцов на возможность преобразования в числовые
# df['HouseStyle']=df['HouseStyle'].str.replace('Story','').astype(float)
# df['HouseStyle'].unique()
categorical_cols = df.select_dtypes(include=['object']) # выбор категориальных колонок
categorical_cols.iloc[:,10:20]
```

```python
df[['LotArea','SalePrice']].describe()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df[['LotArea','SalePrice']],kde=True,color='skyblue')
plt.title('SalePrice')
plt.show
```

```python
# обнаружение и обработка выбросов в столбце SalePrice
# z_scores = np.abs(stats.zscore(df.select_dtypes(include=np.number))) # z оценки для числовых переменных
z_scores = np.abs(stats.zscore(df[['LotArea','SalePrice']])) # z оценки для числовых переменных
df = df[(z_scores <3).all(axis=1)] # удаление строк с выбросами
df.info()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df[['LotArea','SalePrice']],kde=True,color='skyblue')
plt.title('SalePrice')
plt.show
```

```python
# стандартизицая данных (числовых переменных)
df_standardized = df.copy() # копия датафрейм
df_standardized[numeric_cols.columns] = (df_standardized[numeric_cols.columns] - df_standardized[numeric_cols.columns].mean()) / df_standardized[numeric_cols.columns].std()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df_standardized['LotArea'],kde=True,color='skyblue')
plt.title('LotArea')
plt.show
```

```python
plt.figure(figsize=(10, 6)) # Задание размера графика
sns.scatterplot(x='SalePrice', y='OverallQual', data=df) # Создание диаграммы рассеяния
plt.xlabel('SalePrice') # Подпись оси X
plt.ylabel('OverallQual') # Подпись оси Y
plt.title('Relationship between SalePrice and OverallQual') # Название графика
plt.show()
```

```python
label_encoder=LabelEncoder()
df['HouseStyle_Encoded']=label_encoder.fit_transform(df['HouseStyle'])
df[['HouseStyle','HouseStyle_Encoded']].head(30)
```

```python
# one hot encoding
df = pd.get_dummies(df, columns=['RoofStyle'], prefix='RoofStyle', drop_first=False)
```

```python
df.head(10)
```

```python
df['Age']=datetime.now().year-df['YearBuilt']
df[['YearBuilt','Age']]
```

```python
# распределение платных и бесплатных приложений
plt.figure(figsize=(7, 5))
sns.countplot(x="GarageCars", data=df)
plt.title("GarageCars")
plt.show()
```

```python
# создание сводной таблицы
pivot_table = df.pivot_table(index='HouseStyle', columns='BldgType', values='Age', aggfunc='mean')
print(pivot_table)
```
